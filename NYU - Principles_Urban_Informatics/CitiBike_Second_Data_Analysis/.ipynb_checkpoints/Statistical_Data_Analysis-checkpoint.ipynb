### BY GENDER
##### Splitting data by gender and extracting the age happens in the cell below

```python
#df is the dataframe where the content of the csv file is stored
df['ageM'] = 2015 - df['birth year'][(df['usertype'] == 'Subscriber') & (df['gender'] == 1)]
df['ageF'] = 2015 - df['birth year'][(df['usertype'] == 'Subscriber') & (df['gender'] == 2)]
```

```python
df.drop(['tripduration', u'starttime', u'stoptime', u'start station id',
       u'start station name', u'start station latitude',
       u'start station longitude', u'end station id', u'end station name',
       u'end station latitude', u'end station longitude', u'bikeid',
       u'usertype', u'birth year', u'gender', u'date'], axis =1, inplace = True)
df.head()
```

```python
df.describe()
```

```python
df['ageM'].dropna(inplace= True)
df['ageF'].dropna(inplace= True)
df.head()
```

```python
bins = np.arange(10, 99, 5)
axM = df.ageM.groupby(pd.cut(df.ageM, bins)).agg([count_nonzero]).plot(kind='bar', 
                                                                legend=False)
axM.set_title("male riders")
axF = df.ageF.groupby(pd.cut(df.ageF, bins)).agg([count_nonzero]).plot(kind='bar',
                                                                legend=False)
axF.set_title("female riders")
```

### Figure 1: Histogrammed distribution of riders' ages by gender
##### This is how the cumulative distributions look like.  Notice that i am normalizing them! if i want to reat an observed distribution like a probablility distribution i have to normalize it

```python
csM=df.ageM.groupby(pd.cut(df.ageM, bins)).agg([count_nonzero]).cumsum()

csF=df.ageF.groupby(pd.cut(df.ageF, bins)).agg([count_nonzero]).cumsum()

print (np.abs(csM / csM.max()-csF / csF.max()))

pl.plot(bins[:-1] + 5, csM / csM.max(), label = "M")
pl.plot(bins[:-1] + 5, csF / csF.max(), label = "F")
pl.plot(bins[:-1] + 5, np.sqrt(csF / csF.max() - csM / csM.max())**2, 'k-',
        label = "difference")
pl.xlabel("Age")
pl.ylabel("Normalized Cumulative Number")
pl.legend()
```

### Figure 2: The cumulative distribution of CitiBike riders' ages by gender

Lets run the scipy KS test

```python
import scipy.stats
```

### KS tests to compare 2 samples

http://docs.scipy.org/doc/scipy-0.15.1/reference/generated/scipy.stats.ks_2samp.html



```python
ks = scipy.stats.ks_2samp(df.ageM, df.ageF)
print (ks)
```

### The scipy KS test returned basically a p value of 0, or so close to zero that it is zero. 
### Because the P value is way below alpha of 0.05, we can reject the Null Hypothesis ($H_0$)

The scipy.stats KS test already tells me the significance and the p-value. 

The next few cells are here just to show you how you would obtain the same result by hand, but they are **not required**. 

Remember: the Null hypothesis is rejected if 

$D_KS(n1,n2) > c(\alpha) \sqrt{\frac{(n1 + n2)}{n1n2}}$

(see class notes) where $c(\alpha$) is the inverse of the KS distribution, and you do not have to know how to get that cause there are tables that list critical values!! 

http://www.real-statistics.com/tests-normality-and-symmetry/statistical-tests-normality-symmetry/kolmogorov-smirnov-test/kolmogorov-distribution/

But also this result depends in your choice of binning through, and thustheresultyou get by hand may not be exactly the same as the one the KS returns. Either way: this is how you would calculate the KS statistics by hand.

### Now redo the test with a subsample of the data: take only 1 ride every of 200

```python
#ks = scipy.stats.ks_2samp(df.ageM[:200], df.ageF[:200])
ks = scipy.stats.ks_2samp(df.ageM[::200], df.ageF[::200])
print (ks)
```

### The Scipy-KS-Test returned a value that is very close to zero.
### Since the p-value is a lot below the 0.05 mark we can reject the Null Hypothesis ($H_0$)

## Pearson's  test for correlation

#### Notice that the Pearson's is a pairwise test: the samples need to be 
##### a. sorted and 
##### b. the same size!

http://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.pearsonr.html#scipy.stats.pearsonr

```python
Male_sort = df['ageM'].sort_values()
Female_sort = df['ageF'].sort_values()

scipy.stats.pearsonr?

scipy.stats.pearsonr(Male_sort[0:100], Female_sort[0:100])
```

#### Pearson correlation checks the correlation coefficient between two samples and the p-value. 
#### The test assumes that the samples compared are normally distributed (as explained in class)

##### If p-value is > 0.05 fail to reject the null hypothesis and no correlation. 
##### According to studies, it also states that Pearson test gives the output in the range of -1 to 1 so if we get a value closer to 0 instead - that means there is no significant correlation between the samples.

## Spearman's  test for correlation

http://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.spearmanr.html#scipy.stats.spearmanr

```python
scipy.stats.spearmanr?

scipy.stats.spearmanr(Male_sort[0:100], Female_sort[0:100])
```

##### Like Pearson's test, Spearman's test calculates a correlation coefficient and a p-value to test for non-correlation. 
##### The difference is that Spearman's test does not assume normal distribution in the two samples. 
##### The Spearman correlation coefficient ranges from -1 to 1, with 0 indicating no significant correlation between the samples.
